<h2 align="center"><font size="4"> Data Viz </font></h2>

```python
#Como vou análisar apenas a altura em cm, irei remover as duas ultimas colunas
```

```python
df = df[['Rank', 'Country Name', 'Male Height in Cm', 'Female Height in Cm']]
df.head()
```

```python
#Histograma da altura masculina
#Histograma é como uma distribuição de frequência
plt.figure(figsize=(10,5))

male_height = sns.histplot(data=df, x='Male Height in Cm', bins=15, kde=True)
male_height.set_title("Altura Masculina (cm)")
male_height.set_xlabel('Altura (cm)')

plt.show()
```

```python
#Histograma da altura feminina 
plt.figure(figsize=(10,5))

famale_height = sns.histplot(data=df, x='Female Height in Cm', bins=15, kde=True)
famale_height.set_title("Altura Feminina (cm)")
famale_height.set_xlabel('Altura (cm)')

plt.show()
```

#### 1. Altura Masculina

```python
#Atribuindo um novo df com os 10 valores máximos referente a altura masculina
tall_male = df.sort_values(by='Male Height in Cm', ascending=False).head(10)[['Country Name', 'Male Height in Cm']]
```

```python
#Top 10 das alturas masculinas mais altas do mundo
plt.figure(figsize=(20,10))
ax = sns.barplot(x='Country Name', y='Male Height in Cm', data=tall_male, palette='Set1')
ax.set_ylim(180, 185)
ax.set_title('Top 10 Países com Homens mais Altos', pad=10)
ax.set_xlabel("Países")
ax.set_ylabel('')

plt.show()
```

```python
#Atribuindo um novo df com os 10 valores mínimos referente a altura masculina
short_male = df.sort_values(by='Male Height in Cm', ascending=True).head(10)[['Country Name', 'Male Height in Cm']]
```

```python
#Top 10 das alturas masculinas mais baixas do mundo
plt.figure(figsize=(20,10))
ax = sns.barplot(x='Country Name', y='Male Height in Cm', data=short_male, palette='Set1')
ax.set_ylim(159, 166)
ax.set_title('Top 10 Países com Homens mais Baixos', pad=10)
ax.set_xlabel("Países")
ax.set_ylabel('')

plt.show()
```

#### 1. Altura Feminina

```python
#Atribuindo um novo df com os 10 valores máximos referente a altura feminina
tall_female = df.sort_values(by='Female Height in Cm', ascending=False).head(10)[['Country Name', 'Female Height in Cm']]
```

```python
#Top 10 das alturas feminas mais altas do mundo
plt.figure(figsize=(20,10))
ax = sns.barplot(x='Country Name', y='Female Height in Cm', data=tall_female, palette='Set1')
ax.set_ylim(165, 170)
ax.set_title('Top 10 Países com Mulheres mais Altas', pad=10)
ax.set_xlabel("Países")
ax.set_ylabel('')

plt.show()
```

```python
#Atribuindo um novo df com os 10 valores máximos referente a altura feminina
short_female = df.sort_values(by='Female Height in Cm', ascending=True).head(10)[['Country Name', 'Female Height in Cm']]
```

```python
#Top 10 das alturas feminas mais baixas do mundo
plt.figure(figsize=(20,10))
ax = sns.barplot(x='Country Name', y='Female Height in Cm', data=short_female, palette='Set1')
ax.set_ylim(150, 155)
ax.set_title('Top 10 Países com Mulheres mais Baixas', pad=10)
ax.set_xlabel("Países")
ax.set_ylabel('')

plt.show()
```
